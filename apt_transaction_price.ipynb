### 아파트 매매 상세 (기간별)

```python
def get_sigungu_code_arr_by_city_name(city:str = 'ALL') :
    import PublicDataReader as pdr
    df_dong = pdr.code_hdong() # 행정동
    df_sigungu = df_dong[['시도명', '시군구코드', '시군구명']].drop_duplicates(subset=['시도명', '시군구코드', '시군구명'])
    df_sigungu = df_sigungu[df_sigungu['시군구명']!='']
    sgg_cd = []
    if(city.upper() == 'ALL') :
        sgg_cd = df_sigungu['시군구코드']
    else :
        sgg_cd = df_sigungu[df_sigungu['시도명'].str.contains(city)]['시군구코드']
    
    return sgg_cd.values
```

```python
sgg_all = get_sigungu_code_arr_by_city_name()
print(len(sgg_all))
```

```python
def save_trading_data(bs_ym:str = '190001') :
    import os
    import re
    pattern = r'^\d{6}$'

    if (bs_ym == '190001') :
        print('데이터를 저장 할 년월(YYYYMM)을 입력해 주세요')
    elif(re.match(pattern, bs_ym)) :
        api = TransactionPrice(service_key_apt_trading_desc)
        sgg_all = get_sigungu_code_arr_by_city_name()
        # 전월세 데이터 저장
        ssg_tot = len(sgg_all)

        for idx, code_sgg in enumerate(sgg_all) :

            # print(f'[START] 기준월:{bs_ym}, No: {idx+1} / {ssg_tot}, 시군구코드:  {code_sgg}')
            # if(code_sgg >= '11530'):

            df_temp = api.get_data(
                property_type="아파트",
                trade_type="매매",
                sigungu_code=code_sgg,
                year_month=bs_ym,
            )
            file_path = f'/content/drive/MyDrive/Colab Notebooks/apt_transaction_price/data/trading/trading_{bs_ym}.csv'
            # 최초 생성 이후 mode는 append
            if not os.path.exists(file_path):
                df_temp.to_csv(file_path, index=False, mode='w')
            else:
                df_temp.to_csv(file_path, index=False, mode='a', header=False)
            print(f'[END] 기준월:{bs_ym}, No: {idx+1} / {ssg_tot}, 시군구코드:  {code_sgg}')
    else :
        print('정확한 날짜포맷(YYYYMM) 형식으로 입력해 주세요.')
```

### 전월세 데이터 저장

```python
def save_monthly_data(bs_ym:str = '190001') :
    import os
    import re
    pattern = r'^\d{6}$'

    if (bs_ym == '190001') :
        print('데이터를 저장 할 년월(YYYYMM)을 입력해 주세요')
    elif(re.match(pattern, bs_ym)) :
        api_monthly = TransactionPrice(service_key_apt_monthly)
        sgg_all = get_sigungu_code_arr_by_city_name()
        # 전월세 데이터 저장
        ssg_tot = len(sgg_all)
        for idx, code_sgg in enumerate(sgg_all) :
            # print(f'[START] 기준월: {bs_ym}, No: {idx+1} / {ssg_tot}, 시군구코드:  {code_sgg}')

            # if (code_sgg >= '47850') :

            df_temp = api_monthly.get_data(
                property_type="아파트",
                trade_type="전월세",
                sigungu_code=code_sgg,
                year_month=bs_ym,
            )
            file_path = f'/content/drive/MyDrive/Colab Notebooks/apt_transaction_price/data/monthly/monthly_{bs_ym}.csv'
            # 최초 생성 이후 mode는 append
            if not os.path.exists(file_path):
                df_temp.to_csv(file_path, index=False, mode='w')
            else:
                df_temp.to_csv(file_path, index=False, mode='a', header=False)
            print(f'[END] 기준월: {bs_ym}, No: {idx+1} / {ssg_tot}, 시군구코드:  {code_sgg}')
    else :
        print('정확한 날짜포맷(YYYYMM) 형식으로 입력해 주세요.')
```
